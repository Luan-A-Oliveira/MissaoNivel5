```python
from num2words import num2words
from wordcloud import WordCloud
from faker import Faker
import random, ast
import matplotlib.pyplot as plt

#gerar nomes e notas com faker
criar = Faker('pt_BR')
quantidade = 20
notas = 10
lista = {criar.name(): random.randint(0,notas) for i in range(quantidade)}

#gravar e recuperar dados em txt
open('notas.txt', 'w').write(str(lista))   
lista = ast.literal_eval(open('notas.txt', 'r').read())

#manupular os dados
lista = [lista[i] for i in lista]
maior = 0
for i in lista:
    vezes = lista.count(i)
    if vezes > maior:
        maior = vezes

#historiograma dos dados
plt.hist(lista, notas, facecolor='blue', alpha=1, edgecolor='white', linewidth=0.1)
plt.xlabel('pontuações')
plt.ylabel('probabilidade')
plt.title('Histograma das pontuações')
plt.show()

#nuvem de palavras com wordcloud
lista = ','.join([num2words(i, lang='pt') for i in lista])
nuvem_palavras = WordCloud(background_color='black').generate(lista)
plt.imshow(nuvem_palavras)
plt.xticks([])   
plt.yticks([])
```
